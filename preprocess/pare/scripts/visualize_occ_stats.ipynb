```python
import matplotlib.pyplot as plt
import joblib
from collections import Counter
import numpy as np
import pylab
pylab.rcParams['figure.figsize'] = (24.0, 6.0)
pylab.rcParams['font.size'] = 18
```

```python
occ_data = joblib.load('/is/cluster/work/mkocabas/projects/pare/data/coco_train2014_occlusion_stats.pkl')
```

```python
for joint,v in occ_data.items():

    counts = Counter(v)
    
    _counts = counts.copy()
    
    del counts['person']
    
    for k, v in _counts.items():
        if v < 5:
            del counts[k]
    
    labels, values = zip(*counts.items())
    
    # sort your values in descending order
    indSort = np.argsort(values)[::-1]

    # rearrange your data
    labels = np.array(labels)[indSort]
    values = np.array(values)[indSort]

    indexes = np.arange(len(labels))

    bar_width = 0.2
    
    plt.title(f'\"{joint}\" occluders coco_train2014')
    plt.bar(indexes, values)

    # add labels
    plt.xticks(indexes + bar_width, labels, rotation=90)
    plt.show()
```
